# Add Figures and Tables to .ppt

```python
from PIL import Image
import pandas as pd
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor

# Function to add an image to a specific slide
def add_png_to_slide(presentation, slide_number, png_file):
    slide = presentation.slides[slide_number - 1]  # slide_number is 1-based index
    
    # Open the PNG file to get its dimensions
    with Image.open(png_file) as img:
        width, height = img.size
        dpi = img.info.get('dpi', (72, 72))  # Default to 72 dpi if not available
    
    # Convert dimensions from pixels to inches
    width_inch = width / dpi[0]
    height_inch = height / dpi[1]
    
    # Get slide dimensions
    slide_width = presentation.slide_width / 914400  # Convert from EMU to inches
    slide_height = presentation.slide_height / 914400  # Convert from EMU to inches
    
    # Calculate the scaling factor to fit the image within the slide while maintaining aspect ratio
    width_scale = slide_width / width_inch
    height_scale = slide_height / height_inch
    scale = min(width_scale, height_scale)
    
    # Calculate new dimensions
    new_width_inch = width_inch * scale - 0.25
    new_height_inch = height_inch * scale - 0.25
    
    # Calculate position to center the image
    left_inch = (slide_width - new_width_inch) / 2
    top_inch = (slide_height - new_height_inch) / 2
    
    # Add picture to slide
    pic = slide.shapes.add_picture(png_file, Inches(left_inch), Inches(top_inch), Inches(new_width_inch), Inches(new_height_inch))
    return pic

# Function to add a table to a specific slide
def add_table_to_slide(presentation, slide_number, df, left, top, width, height):
    slide = presentation.slides[slide_number - 1]  # slide_number is 1-based index
    
    # Add a table to the slide
    rows, cols = df.shape
    table = slide.shapes.add_table(rows + 1, cols, Inches(left), Inches(top), Inches(width), Inches(height)).table
    
    # Set column widths
    for i, width in enumerate([1.5] * cols):  # Adjust column widths as needed
        table.columns[i].width = Inches(width)
    
    # Set row heights
    for i in range(rows + 1):
        table.rows[i].height = Inches(0.25)  # Adjust row height as needed
    
    # Write DataFrame to table
    for col_num, column_title in enumerate(df.columns):
        cell = table.cell(0, col_num)
        cell.text = column_title
        cell.text_frame.paragraphs[0].font.bold = True
        cell.text_frame.paragraphs[0].font.size = Pt(8)  # Set font size to 8
        cell.text_frame.paragraphs[0].alignment = PP_ALIGN.CENTER
        cell.fill.solid()
        cell.fill.fore_color.rgb = RGBColor(0xD9, 0xD9, 0xD9)  # Gray background
    
    for row_num in range(rows):
        for col_num in range(cols):
            cell = table.cell(row_num + 1, col_num)
            cell.text = str(df.iat[row_num, col_num])
            cell.text_frame.paragraphs[0].font.size = Pt(8)  # Set font size to 8
            cell.text_frame.paragraphs[0].alignment = PP_ALIGN.CENTER
    
    return table

# Load the existing presentation
ppt = Presentation("Slides/2022_COWER_Final_Version.pptx")

# Define DataFrames and their titles
dataframes_info = [
    {"file": "Tables/Summary_Table_Wind_ES.csv", "title": "Executive Summary Table", "slide_number": 6},
    {"file": "Tables/LBW_CapEx.csv", "title": "Land Based Wind CapEx", "slide_number": 27},
    {"file": "Tables/LBW_OpEx.csv", "title": "Land Based Wind OpEx", "slide_number": 29},
    {"file": "Tables/Summary_Table_LBW.csv", "title": "Land Based Wind", "slide_number": 31},
    {"file": "Tables/FBOW_CapEx.csv", "title": "Fixed-Bottom Offshore Wind CapEx", "slide_number": 37},
    {"file": "Tables/FLOW_CapEx.csv", "title": "Floating Offshore Wind CapEx", "slide_number": 38},
    {"file": "Tables/OW_OpEx.csv", "title": "Offshore Wind OpEx", "slide_number": 39},
    {"file": "Tables/Summary_Table_OSW.csv", "title": "Offshore Wind", "slide_number": 41},
    {"file": "Tables/DW_CapEx.csv", "title": "Distributed Wind LCOE", "slide_number": 49}
]

# Add each DataFrame as a table to the specified slide
for info in dataframes_info:
    df = pd.read_csv(info["file"])
    add_table_to_slide(ppt, info["slide_number"], df, left=1, top=1, width=4, height=4)  # Adjust dimensions as needed

# Define the PNG files and corresponding slide numbers
figures_info = [
    #land based plots
    {"file": "Figures/LBW_LCOE_waterfall.png", "slide_number": 7},
    {"file": "Figures/GPRA_LBW.png", "slide_number": 22},
    {"file": "Figures/LBW_capex_donut.png", "slide_number": 27},
    {"file": "Figures/LBW_LCOE_waterfall.png", "slide_number": 32},
    {"file": "Figures/LBW_LCOE_sensitivity_percentage.png", "slide_number": 33},
    #fixed-bottom offshore wind plots
    {"file": "Figures/FBOW_LCOE_waterfall.png", "slide_number": 8},
    {"file": "Figures/GPRA_FBOW.png", "slide_number": 23},
    {"file": "Figures/FBOW_capex_donut.png", "slide_number": 37},
    {"file": "Figures/FBOW_LCOE_waterfall.png", "slide_number": 42},
    {"file": "Figures/FBOW_LCOE_sensitivity_percentage.png", "slide_number": 44},
    #floating offshore wind plots
    {"file": "Figures/FLOW_LCOE_waterfall.png", "slide_number": 9},
    {"file": "Figures/FLOW_capex_donut.png", "slide_number": 38},
    {"file": "Figures/FLOW_LCOE_waterfall.png", "slide_number": 43},
    {"file": "Figures/FLOW_LCOE_sensitivity_percentage.png", "slide_number": 45},
    #distributed wind plots
    {"file": "Figures/DW_20kW_LCOE_waterfall.png", "slide_number": 10},
    {"file": "Figures/DW_100kW_LCOE_waterfall.png", "slide_number": 10},
    {"file": "Figures/DW_1500kW_LCOE_waterfall.png", "slide_number": 10},
    {"file": "Figures/DW_20kW_capex_donut.png", "slide_number": 49},
    {"file": "Figures/DW_100kW_capex_donut.png", "slide_number": 49},
    {"file": "Figures/DW_1500kW_capex_donut.png", "slide_number": 49},
    {"file": "Figures/DW_20kW_LCOE_waterfall.png", "slide_number": 51},
    {"file": "Figures/DW_100kW_LCOE_waterfall.png", "slide_number": 51},
    {"file": "Figures/DW_1500kW_LCOE_waterfall.png", "slide_number": 51},
]

# Add each figure to the specified slide
for figure in figures_info:
    add_png_to_slide(ppt, figure["slide_number"], figure["file"])

# Save the presentation
ppt.save("Slides/2023_COWER_Working_Version.pptx")
```

# Update numbers from Slide 11 (Key Conclusions)

```python
import pandas as pd
from pptx import Presentation

# Function to format numbers with commas for thousands
def format_number(value):
    return "{:,}".format(int(value))

# Function to replace text in a specific slide
def replace_text_in_slide(slide, replacements):
    for shape in slide.shapes:
        if not shape.has_text_frame:
            continue
        text_frame = shape.text_frame
        for paragraph in text_frame.paragraphs:
            for run in paragraph.runs:
                text = run.text
                for key, value in replacements.items():
                    if key in text:  # Check if the text contains the key
                        text = text.replace(key, value)  # Replace text
                run.text = text

# Load the existing presentation
ppt = Presentation("Slides/2023_COWER_Working_Version.pptx")

# Read the values from the CSV file for text replacement
df = pd.read_csv("Tables/Summary_Table_Wind_ES.csv")

# Extract relevant values from the DataFrame based on row indices
# Assume the row for LCOE values is at index 6 (adjust if necessary)
lcoe_row = df.iloc[6]


# Extract the values from the Summary Table
land_based_lcoe = lcoe_row['Land-Based']
fixed_bottom_lcoe = lcoe_row['Offshore']
floating_lcoe = lcoe_row['Offshore.1']
single_turbine_residential_lcoe = lcoe_row['Distributed']
single_turbine_commercial_lcoe = lcoe_row['Distributed.1']
single_turbine_large_lcoe = lcoe_row['Distributed.2']

# Read the sensitivity CSV files
lbw_sensitivity_df = pd.read_csv("Data/LBW_Sensitivity.csv")
fbow_sensitivity_df = pd.read_csv("Data/FBOW_Sensitivity.csv")
flow_sensitivity_df = pd.read_csv("Data/FLOW_Sensitivity.csv")


# Extract min and max LCOE values
min_lcoe_lbw = lbw_sensitivity_df['min LCOE'].min()  # Update this if the column name is different
max_lcoe_lbw = lbw_sensitivity_df['max LCOE'].max()  # Update this if the column name is different
min_lcoe_fbow = fbow_sensitivity_df['min LCOE'].min()  # Update this if the column name is different
max_lcoe_flow = flow_sensitivity_df['max LCOE'].max()  # Update this if the column name is different

# Extract min and max CapEx values from the first row of each sensitivity CSV
min_capex_fbow = fbow_sensitivity_df['min value'][0]  # Update this if the column name is different
max_capex_flow = flow_sensitivity_df['max value'][0]  # Update this if the column name is different

# Format extracted values
land_based_lcoe_str = format_number(land_based_lcoe)
fixed_bottom_lcoe_str = format_number(fixed_bottom_lcoe)
floating_lcoe_str = format_number(floating_lcoe)
single_turbine_residential_lcoe_str = format_number(single_turbine_residential_lcoe)
single_turbine_commercial_lcoe_str = format_number(single_turbine_commercial_lcoe)
single_turbine_large_lcoe_str = format_number(single_turbine_large_lcoe)
min_lcoe_lbw_str = format_number(min_lcoe_lbw)
max_lcoe_lbw_str = format_number(max_lcoe_lbw)
min_lcoe_fbow_str = format_number(min_lcoe_fbow)
max_lcoe_flow_str = format_number(max_lcoe_flow)
min_capex_fbow_str = format_number(min_capex_fbow)
max_capex_flow_str = format_number(max_capex_flow)

# Prepare replacements dictionary
replacements = {
    "39": land_based_lcoe_str,  # Land-Based LCOE
    "95": fixed_bottom_lcoe_str,  # Fixed-Bottom LCOE
    "145": floating_lcoe_str,  # Fixed-Bottom LCOE
    "235": single_turbine_residential_lcoe_str,  # Single Turbine (Residential) LCOE
    "163": single_turbine_commercial_lcoe_str,  # Single Turbine (Commercial) LCOE
    "78": single_turbine_large_lcoe_str,  # Single Turbine (Large) LCOE
    "30": min_lcoe_lbw_str,  # Minimum LCOE from LBW Sensitivity
    "57": max_lcoe_lbw_str,  # Maximum LCOE from LBW Sensitivity
    "52": min_lcoe_fbow_str,  # Minimum LCOE from FBOW Sensitivity
    "184": max_lcoe_flow_str,  # Maximum LCOE from FLOW Sensitivity
    "1,800": min_capex_fbow_str,  # Minimum CapEx from FBOW Sensitivity
    "7,711": max_capex_flow_str  # Maximum CapEx from FLOW Sensitivity
}

# Specify the slide number where text needs to be replaced
slide_number = 11
slide = ppt.slides[slide_number - 1]

# Replace text in the slide
replace_text_in_slide(slide, replacements)

# Specify the slide number where text needs to be replaced
slide_number = 41
slide = ppt.slides[slide_number - 1]

# Replace text in the slide
replace_text_in_slide(slide, replacements)

# Save the updated presentation
ppt.save("Slides/2023_COWER_Working_Version.pptx")
```
